## İLERİ TOPLULAŞTIRMA İŞLEMLERİ (AGGREGATE, FİLTER)

## aggregate
#### TANIMLAMA FONKSİYONU İLE TAMAMINI ALMAK YERİNE BELİRLİ VE SEÇTİĞİMİZ FONKSİYOONLARIN ÇALIŞTIRIILMASINI SAĞLAR

```python
df
```

```python
# Her bir Özellik için tek tek istediğim değerleri çıkartarak yazdı.
df.groupby("method").aggregate(["min", "median","max"])
```

```python
# iki farklı özellik için farklı değerleri almak istersem... 
# Birinin max ı, birinin min i gibi bunu da dict ile yapıyoruz... 
df.groupby("method").aggregate({"number":"min", "mass":"max"})
```

# CONDITIONAL FILTER 

```python
def std_filter(x):
    return x["mass"].std() > 1
```

```python
# BİR FONKSİYONU PARAMETRE OLARAK ATARKEN PARANTEZ KULLANILMAZ.

df.groupby("method").filter(std_filter)
```

# PIVOT TABLOLAR

```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.groupby("sex")["survived"].mean() #survived tek parantexle de kullanılır, iki parantex detaylandırır
```

```python
titanic.groupby("sex")[["survived"]].mean() #DataFrame şeklinde gelir
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean")
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate(["mean","std"])
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate(["mean","std"]).unstack()
```
